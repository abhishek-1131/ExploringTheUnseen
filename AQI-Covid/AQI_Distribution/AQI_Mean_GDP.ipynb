```python
import pandas as pd
import numpy as np
import plotly.express as px
import datetime
import matplotlib.pyplot as plt
import plotly.figure_factory as ff
import plotly.graph_objects as go
import scipy.stats as stats
import math
```

```python
df19=pd.read_csv('2019_7cities.csv')
df20=pd.read_csv('2020_7cities.csv')

datel=[]
format_str = '%d/%m/%Y'

startdate = pd.to_datetime('2019-03-21').date()
enddate = pd.to_datetime('2019-06-28').date()

for i in range(df19.shape[0]):
    datetime_obj = datetime.datetime.strptime(df19['Date'].iloc[i], format_str)
    datel.append(datetime_obj.date())
df19['Date']=datel
df19 = df19[(df19['Date'] > startdate) & (df19['Date'] <= enddate)]

datel=[]
startdate = pd.to_datetime('2020-03-21').date()
enddate = pd.to_datetime('2020-06-28').date()
for i in range(df20.shape[0]):
    datetime_obj = datetime.datetime.strptime(df20['Date'].iloc[i], format_str)
    datel.append(datetime_obj.date())
datel
df20['Date']=datel
df20 = df20[(df20['Date'] > startdate) & (df20['Date'] <= enddate)]


city_list1= ['Mumbai','Kolkata','Bengaluru','Hyderabad','Visakhapatnam','Delhi']
med=[]
df19=df19[df19['Specie'].isin(['pm25'])]
for i in city_list1:
    df1=df19[df19.City == i]
    x=df1['median'].to_numpy()
    med.append(x)

hist_data=[med[0],med[1],med[2],med[3],med[4],med[5]]
fig = ff.create_distplot(hist_data,city_list1, bin_size=5,show_hist=False)
fig.show()
med=[]
df20=df20[df20['Specie'].isin(['pm25'])]
for i in city_list1:
    df2=df20[df20.City == i]
    x=df2['median'].to_numpy()
    med.append(x)
hist_data=[med[0],med[1],med[2],med[3],med[4],med[5]]
fig = ff.create_distplot(hist_data,city_list1, bin_size=5,show_hist=False)
fig.show()
```
