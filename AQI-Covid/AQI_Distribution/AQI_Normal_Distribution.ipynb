```python
df=pd.read_csv('2019_7cities.csv')
city_list= ['Mumbai','Kolkata','Bengaluru','Chennai','Hyderabad','Visakhapatnam','Delhi',]
pollutant= ['co','so2','no2','o3','pm10','pm25']
df=df[df['Specie'].isin(['pm25'])]
dff=df
mean19=[]
var19=[]
s=df.shape
for i in city_list:
    df1=df[df.City == i]
    x,y=(df1['median'].mean(),df1['median'].var())
    mean19.append(x)
    var19.append(y)
#mean
#var
```

```python
df=pd.read_csv('2020_7cities.csv')
city_list= ['Mumbai','Kolkata','Bengaluru','Chennai','Hyderabad','Visakhapatnam','Delhi',]
pollutant= ['co','so2','no2','o3','pm10','pm25']
df=df[df['Specie'].isin(['pm25'])]
dff=df
mean20=[]
var20=[]
s=df.shape
for i in city_list:
    df1=df[df.City == i]
    x,y=(df1['median'].mean(),df1['median'].var())
    mean20.append(x)
    var20.append(y)
#mean
#var
```

```python
color=['red','blue','green','black','cyan','brown','purple']
city_list= ['Mumbai','Kolkata','Bengaluru','Chennai','Hyderabad','Visakhapatnam','Delhi',]
for i in range(7):
    mu = mean19[i]
    variance = var19[i]
    sigma = math.sqrt(variance)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, stats.norm.pdf(x, mu, sigma),color=color[i])
    #plt.legend(loc='upper left', frameon=False)
plt.show()
```
