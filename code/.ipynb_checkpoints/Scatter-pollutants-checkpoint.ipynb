```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('seaborn-bright')
```

```python
year = 2020
df = pd.read_csv(f'data/{year}_7cities.csv')
```

```python
df.Date = pd.to_datetime(df.Date, format = '%d/%m/%Y') # need the format string to specify which is the month and which is the year
df['month_num'] = pd.DatetimeIndex(df['Date']).month
df['month_name'] = df.Date.dt.month_name()
df.dtypes
```

```python
print(df.shape)
# just keep 2019 data, discard everything else
df = df[df['Date'].dt.year == year]
print(df.shape)
```

```python
cities = df.City.unique()
pollutants = df.Specie.unique()

print(cities)
print(pollutants)
print(df.head())
```

```python
df_s = df[df.City == 'Delhi']
df_s_1 = df_s.loc[df_s.Specie == 'pm25', ['Date','median']]
df_s_1.rename(columns = {'median':'pm25'}, inplace=True)
df_s_1.set_index('Date', inplace=True)

df_s_2 = df_s.loc[df_s.Specie == 'pm10', ['Date','median']]
df_s_2.rename(columns = {'median':'pm10'}, inplace=True)
df_s_2.set_index('Date', inplace=True)

df_c = pd.concat([df_s_1, df_s_2], axis=1)

print(df_s_1.shape)
print(df_s_1.head())
print(df_s_2.shape)
print(df_s_2.head())
print(df_c.head())
```

```python
print(len(df_s_1.index.duplicated()), np.count_nonzero(df_s_1.index.duplicated()))
print(df_s_1.tail(10))
#print(df_s_1.index.duplicated())
```

```python
df_c.plot.scatter('pm25','pm10')
plt.show()
```

```python
fig, ax = plt.subplots(1, len(cities), figsize = (24, 8))
for i, c in enumerate(cities):
    df_s = df[df.City == c]
    df_s_pm25 = df_s.loc[df_s.Specie == 'pm25', ['Date','median']]
    df_s_pm25.rename(columns = {'median':'pm25'}, inplace=True)
    df_s_pm25.set_index('Date', inplace=True)
    

    df_s_pm10 = df_s.loc[df_s.Specie == 'pm10', ['Date','median']]
    df_s_pm10.rename(columns = {'median':'pm10'}, inplace=True)
    df_s_pm10.set_index('Date', inplace=True)

    df_c = pd.concat([df_s_pm25, df_s_pm10], axis=1)
    df_c['month_num'] = pd.DatetimeIndex(df_c.index.month)
    df_c.plot.scatter(x='pm25',y='pm10', c=df_c['month_num'], ax=ax[i], colormap='winter')
 ## todo: remove all except the last colorbar 
    #df_c.plot.scatter('pm25','pm10', ax=ax[i])
    ax[i].set_ylim(bottom=0)
    ax[i].set_xlim(left=0)
    ax[i].set_title(c)
    print(c)
    print(df_c.corr())
plt.show()
    

    
```
