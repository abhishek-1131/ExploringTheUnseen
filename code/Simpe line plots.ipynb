## Simple Plots

Lets import some Libraries

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
pio.templates.default = "plotly"
```

```python
df=pd.read_csv('data/2020-28thJuly_7cities.csv')
df.head(5)
```

Lets see the Cities and the pollutants

```python
print(df['City'].unique())
print(df['Specie'].unique())
```

Focusing on PM 2.5

```python
gas=['pm25']
df1=df[df['Specie'].isin(gas)]
df1=df1.reset_index()
```

Lets make a simple bar chart

```python
fig=px.bar(df1,'Date','median',color='City')
fig.update_layout(title="PM 2.5 for Indian cities")
fig.show()
fig.write_image("pm2.5_barchart.png")
```

```python
fig=px.line(df1,x='Date',y='median',color='City')
fig.add_shape(type="line",
            x0='30/12/2019',
            y0=25,
            x1='25/07/2020',
            y1=25,
            line=dict(
                color="orange",
                width=3,
                dash="dot",
            ))
fig.add_shape(type="line",
            x0='30/12/2019',
            y0=60,
            x1='25/07/2020',
            y1=60,
            line=dict(
                color="black",
                width=3,
                dash="dot",
            ))
fig.update_layout(title="PM 2.5 for Indian cities")
fig.show()
```

Too crowded, right? Lets do it for fewer cities.

```python
city=['Mumbai','Delhi','Kolkata']
dfc=df1[df1['City'].isin(city)]
```

```python
fig=px.line(dfc,x='Date',y='median',color='City')
fig.add_shape(type="line",
            x0='30/12/2019',
            y0=25,
            x1='25/07/2020',
            y1=25,
            line=dict(
                color="orange",
                width=3,
                dash="dot",
            ))
fig.add_shape(type="line",
            x0='30/12/2019',
            y0=60,
            x1='25/07/2020',
            y1=60,
            line=dict(
                color="green",
                width=3,
                dash="dot",
            ))
fig.show()
fig.write_image('pm2.5_line_chart_3.png')
```

Lets focus on Mumbai

```python
dfc=df1[df1['City'].isin(['Mumbai'])]
fig=px.line(dfc,x='Date',y='median',color='City')
fig.add_shape(type="line",
            x0='30/12/2019',
            y0=25,
            x1='25/07/2020',
            y1=25,
            line=dict(
                color="orange",
                width=3,
                dash="dot",
            ))
fig.add_shape(type="line",
            x0='30/12/2019',
            y0=60,
            x1='25/07/2020',
            y1=60,
            line=dict(
                color="green",
                width=3,
                dash="dot",
            ))
fig.update_layout(title="PM 2.5 - Mumbai",showlegend=False)
fig.show()
fig.write_image('pm2.5_mumbai.png')
```

Lets generate PM 2.5 for all cities 

```python
city=['Delhi','Hyderabad','Bengaluru','Chennai','Mumbai','Kolkata']
p=[1,1,1,2,2,2]
h=[1,2,3,1,2,3]
fig = make_subplots(
rows=2, cols=3,
subplot_titles=("Delhi", "Hyderabad", "Bengaluru", "Chennai",'Mumbai','Kolkata'))
for l,i in enumerate(city):
    dff=df1[df1['City'].isin([i])]
    fig.add_trace(go.Scatter(x=dff['Date'], y=dff['median']),row=p[l], col=h[l])
fig.update_layout(height=650, width=1000,showlegend=False,
                  title_text="PM25 for various cities")
fig.update_layout(yaxis=dict(range=[0,300]))
fig.show()
fig.write_image('pm2.5_all_cities.png')
```

Plot them all!

```python
city=['Delhi','Hyderabad','Bengaluru','Chennai','Mumbai','Kolkata']
pollutant=['pm25','pm10','co','so2','o3','no2']
titles=[]
for i in city:
    for j in pollutant:
        entry= i + '-' + j
        titles.append(entry)
fig = make_subplots(
rows=6, cols=6,
subplot_titles=titles)
for m,i in enumerate(city):
    for n,j in enumerate(pollutant):
        dff=df[df['City'].isin([i])]
        dff=dff[dff['Specie'].isin([j])]
        fig.add_trace(go.Scatter(x=dff['Date'], y=dff['median']),row=m+1, col=n+1)
fig.update_layout(height=1450, width=1400,showlegend=False,
                  title_text="All pollutants for various cities")
fig.show()        
fig.write_image('all_pollutants.png')
```
